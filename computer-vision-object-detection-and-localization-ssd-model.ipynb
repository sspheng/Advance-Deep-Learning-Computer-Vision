## Frozen Model:

A frozen model typically refers to a pre-trained neural network model that has been serialized and optimized for inference. It is usually represented by a file with a specific file format, such as a TensorFlow SavedModel (.pb) or a TensorFlow Lite model (.tflite).
The frozen model contains the network architecture and learned parameters (weights and biases) that were obtained during the training process.
It is called a "frozen" model because the parameters are fixed and not updated during inference, ensuring consistent predictions.
The frozen model file is used as an input to load the pre-trained model into memory during the initialization of the object detection model.




## Configuration File:

The configuration file, often referred to as a "config file," provides additional information about the model architecture and settings required for using the pre-trained model.

It is a separate file typically in a specific format, such as a text-based file or a YAML file, that defines the network structure, input size, output layers, class labels, and other necessary parameters.

The configuration file specifies details like the type of model (e.g., Single Shot Multibox Detector (SSD), You Only Look Once (YOLO), Faster R-CNN, etc.), the number of classes the model can detect, the size of the input images or video frames expected by the model, and any specific preprocessing steps required.


The configuration file is used to configure the object detection model and ensure compatibility with the pre-trained frozen model.

## Frozen Model (ssd_mobilenet_v3_large_coco_2020_01_14.pb):

This refers to a pre-trained object detection model based on the MobileNetV3 architecture.
"ssd_mobilenet_v3_large_coco_2020_01_14.pb" represents the frozen model file saved in the Protocol Buffers format (.pb).

pbtxt File (ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt):

    
The ".pbtxt" file is a text-based file that accompanies the frozen model file (.pb).
It contains the human-readable version of the model's configuration or graph definition in the Protocol Buffers text format.
The pbtxt file provides details about the model's structure, including the network architecture, input size, output layers, and other parameters.
It serves as a companion file to help understand and interpret the frozen model.


The pbtxt file is typically used for visualization, inspection, and customization of the model's behavior.

```python
cap = cv2.VideoCapture("pexels-tim-samuel-5834557-3840x2160-24fps.mp4")

#    "pexels-george-morina-5644228-3840x2160-30fps.mp4")

#if not cap.isOpened():
 #   cap  = cv2.VideoCapture(0) # from local camera


if not cap.isOpened():
    raise IOError("can't open the video")

    
    
while(True):
      
    # Capture the video frame
    # by frame
    ret, frame = cap.read()
    classIndex, confidence, bbox = model.detect(frame, confThreshold=0.5)
    
    print(classIndex)
    
    
    if (len(classIndex)!=0):
        for ClassInd, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox ):
            if (ClassInd<=80):
                cv2.rectangle(frame, boxes, (255,0,0),2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10, boxes[1]+40), font,fontScale= font_scale, color=(0,255,0) #BGR
                       #,thic
                       )

    # Display the resulting frame
    cv2.imshow('Object', frame)
      
    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choice
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
cap.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
