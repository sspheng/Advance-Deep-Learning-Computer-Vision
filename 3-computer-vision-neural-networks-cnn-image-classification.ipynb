# KERAS/TF MODELLING



```python


from numpy import mean
from numpy import std
from matplotlib import pyplot

#from sklearn.model_selection import KFold
```

```python


from keras.datasets import mnist

from keras.utils import to_categorical

from keras.models import Sequential

from keras.layers import Conv2D
from keras.layers import MaxPooling2D

from keras.layers import Flatten
from keras.layers import BatchNormalization

from keras.layers import Dense

#ADAM as well
from keras.optimizers import SGD

```

```python

    (trainX, trainY), (testX, testY) = mnist.load_data()
    
    

    

```

```python
trainX[0]
```

```python
# reshape dataset to have a single channel
    
trainX = trainX.reshape((trainX.shape[0], 28, 28, 1))
    
testX = testX.reshape((testX.shape[0], 28, 28, 1))
    
   
```

```python
trainX[0]
```

```python
trainY
```

```python
 
# one hot encode target values
trainY = to_categorical(trainY)

testY = to_categorical(testY)
```

# scale pixels

```python


    
# convert from integers to floats


train_norm = trainX.astype('float32')
test_norm = testX.astype('float32')
    
    # normalize to range 0-1
   # ------>  x-min/max
    #------>  x- mean/sd
    
train_norm = train_norm / 255.0
test_norm = test_norm / 255.0
    
     # return normalized images
   
 
```

```python
train_norm[0]

```

```python

# define CNN model

def define_model():
    
    model = Sequential()
    
    #conv layer with 32 filters of 3*3 
    
    #try with 5 filters
    
    model.add(Conv2D(32, (3, 3), activation='relu', 
                     #strides = (1,1),
                     #padding="same",
                     kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
    
    #filter -- then -- feature map -- Relu --- output for pooling layer
    
    #the kernel size in Keras,
    #It is common to use 3×3 sized filters, and perhaps 5×5 or even 7×7 sized filters, for larger input images.

    
    #For example, the stride can be changed to (2,2). This has the effect of moving the filter two pixels right for each horizontal movement of the filter and two pixels down for each vertical movement of the filter when creating the feature map.
    #padding

   
    
    #optional
   
    # model.add(BatchNormalization())
    
    
    
    
    #pooling layer  -- MAX and AVERAGE pooling?
    
    model.add(MaxPooling2D((2, 2))) # we can pass stride as well--overlapping vs non-overlapping
    
    
    #mandatory
    #2D or 3D to 1D

    model.add(Flatten())
    
    
    
    #try 20 neurons
    model.add(Dense(80, activation='relu', kernel_initializer='he_uniform'))
    
    #optional
    # model.add(BatchNormalization())
    
    
    ###  10 because numbers to classify are from 0 to 9
    
    model.add(Dense(10, activation='softmax'))
    
    
    # compile model
    
    #TRY ADAM as well
    opt = SGD(lr=0.01, momentum=0.9)
    
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model

```

```python
mdl = define_model()


mdl.summary()
```

```python

mdl.fit(train_norm, trainY, epochs=10, batch_size=32, verbose=1)
    
    
```

```python
loss, acc = mdl.evaluate(test_norm, testY, verbose=1)


```

```python
print('Train Loss is: %.3f' % (loss))

print('Train Accuracy is: %.3f' % (acc * 100.0))

```
