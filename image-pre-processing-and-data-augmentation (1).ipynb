# DATA AUGMENTATION


```python
# convert to gray scale
```

```python
gray_hadoop = cv2.cvtColor(img_hadoop, cv2.COLOR_BGR2GRAY)

gray_hadoop
```

# cv2.cvtColor() method 

* is used to convert an image from one color space to another.
* There are more than 150 color-space conversion methods available in OpenCV. 
* will use some of color space conversion codes below.

```python
color_hadoop = cv2.cvtColor(gray_hadoop, cv2.COLOR_GRAY2RGBA)

color_hadoop
```

```python
cv2.imshow("Gray Hadoop Image", color_hadoop)

cv2.waitKey(0)
```

```python
# rotate an image
# Rotate gray scale image
```

```python
img_gray_hadoop_rot = cv2.rotate(gray_hadoop, cv2.ROTATE_90_CLOCKWISE)

img_gray_hadoop_rot
```

```python


cv2.imshow("Gray Hadoop Rotated Image", img_gray_hadoop_rot)

cv2.waitKey(0)
```

```python
img_gray_hadoop_rot
```

```python
#write image to a file

cv2.imwrite("black_hadoop.jpg", img_gray_hadoop_rot)
```
